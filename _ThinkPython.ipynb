```python
# Ex. 6.11.3 - Palindromes
def is_palindrome(s):
    len_s = len(s)
    if len_s % 2 == 0:
        left_s = s[0:int(len_s/2)]
        right_s = s[int(len_s/2):]
        print("Even")
    else:
        left_s = s[0:int(len_s//2)]
        right_s = s[int(len_s//2+1):]
        print("Odd")
    print(s, len_s, left_s, right_s)
    if left_s == right_s[::-1]:
        return True
    else:
        return False
    
s = input("Enter string: ")
if is_palindrome(s):
    print("%s is a palindrome" % (s))
else:
    print("%s is NOT a palindrome" % (s))
```

```python
# Ex 7.9.1 - sqrt comparisons
import math
def mysqrt(a):
    """Uses Newton's Method to find the square root of a"""
    x = a / 2     # first estimate
    while True:
        y = (x + a/x) / 2
        if abs(y - x) < .0001:
            return(y)
        else:
            x = y

a = 0
print("a", "mysqrt(a)", "math.sqrt(a)", "diff")
print("-", "---------", "------------", "----")
while a < 10:
    a += 1
    m1 = mysqrt(a)
    m2 = math.sqrt(a)
    diff = abs(m1 - m2)
    print(a, m1, m2, diff)
```

```python
# Ex. 7.9.3 Estimate pi
import math
def factorial(n):
    """Computes n! recursively"""
    if n == 0:
        return 1
    else:
        recurse = factorial(n - 1)
        result = n * recurse
        return result
    
def estimate_pi():
    """Estimate the value of pi using algorithm from Srinivasa Ramanujan
    http://en.wikipedia.org/wiki/Pi"""
    factor = 2 * math.sqrt(2) / 9801
    k = 0
    total = 0
    while True:
        num = factorial(4 * k) * (1103 + 26390 * k)
        den = (factorial(k) ** 4) * (396 ** (4 * k))
        term = factor * num / den
        total += term
        if abs(term) < 1e-15:
            break
        k += 1
    return 1 / total
        
pi_est = estimate_pi()
diff = abs(pi_est - math.pi)
print(pi_est, math.pi, diff)
```

```python
"""This module contains a code example related to

Think Python, 2nd Edition
by Allen Downey
http://thinkpython2.com

Copyright 2015 Allen Downey

License: http://creativecommons.org/licenses/by/4.0/
"""

from __future__ import print_function, division


def rotate_letter(letter, n):
    """Rotates a letter by n places.  Does not change other chars.

    letter: single-letter string
    n: int

    Returns: single-letter string
    """
    if letter.isupper():
        start = ord('A')
    elif letter.islower():
        start = ord('a')
    else:
        return letter

    c = ord(letter) - start
    i = (c + n) % 26 + start
    return chr(i)


def rotate_word(word, n):
    """Rotates a word by n places.

    word: string
    n: integer

    Returns: string
    """
    res = ''
    for letter in word:
        res += rotate_letter(letter, n)
    return res


if __name__ == '__main__':
    print(rotate_word('cheer', 7))
    print(rotate_word('melon', -10))
    print(rotate_word('sleep', 9))
```

```python
# Ex. 9.7.7 Puzzler - words with 3 consecutive double letters
def consec_doubles(word):
    """returns True is finds 3 consecutive double letters in word"""
    len_w = len(word)
    if len_w <= 6:
        return False
    i = 0
    while i <= len_w - 6:
        if word[i] == word[i+1]:
            if word[i+2] == word[i+3]:
                if word[i+4] == word[i+5]:
                    return True
        i += 1
    return False

fin = open('words.txt')
count_all = 0
count_3doubles = 0
for line in fin:
    count_all += 1
    word = line.strip()
    if consec_doubles(word):
        count_3doubles += 1
        print(count_3doubles, word)
print("Done. 3-double words: %d out of %d total read." % (count_3doubles, count_all))
```

```python
# 10.7 Ex. 1 - nested_sum
def nested_sum(nested_list):
    """sums all the nested lists of integers"""
    len_list = len(nested_list)
    i = 0
    tot = 0
    while i < len_list:
        if type(nested_list[i]) == int:
            tot += nested_list[i]
        elif type(nested_list[i]) == list:
            tot += nested_sum(nested_list[i])
        i += 1
    return tot

my_list = [1, [2, 3], 4, [6, 7], 18]
print(nested_sum(my_list))
```

```python
# 10.7 Ex. 2 - capitaliz_all with nested list
def capitalize_all(t):
    res = []
    for s in t:
        if type(s) == str:
            res.append(s.capitalize())
        elif type(s) == list:
            res.append(capitalize_all(s))
    return res

my_list = ['all', 'fred', ['judy', 'sue'], 'arabella']
print(capitalize_all(my_list))
```

```python
# 10.7 Ex. 3 - cumulative sum
def cum_sum(list):
    """takes a list of numbers and returns the cumulative sum;
    that is, a new list where the ith element is the sum of the first i+1 elements
    from the original list. For example, the cumulative sum of [1, 2, 3] is [1, 3, 6]. """
    cum = 0
    new_list = []
    for e in list:
        cum += e
        new_list.append(cum)
    return new_list

my_list = [2,34,4,57,11]
print("Original list: ", my_list)
print("Cumulative list: ", cum_sum(my_list))
```

```python
# 10.15 Ex. 8 - Birthday Paradox
import random
def has_duplicates(t):
    """Returns True if an element of list t appears more than once"""
    d = sorted(t)
    i = 0
    while i < len(d) - 1:
        if d[i] == d[i+1]:
            return True
        i += 1
    return False

def gen_list(t):
    """generate list of 23 random ints from 1 to 365 inclusive"""
    for i in range(23):
        t.append(random.randint(1, 365))
    return t

c_tot = 0
c_true = 0
num_tries = 1000
for j in range(num_tries):
    my_list = []
    gen_list(my_list)
    if has_duplicates(my_list):
        c_true += 1
    c_tot += 1
print("%d matches in %d tries, or %g percent." % (c_true, c_tot, 100*c_true/c_tot))
```

```python
# 10.15 Ex. 9 - remove_duplicates
def remove_dups(t):
    """returns new list scrubbed of duplicates"""
    d = sorted(t)
    e = []
    prev = ''
    len_d = len(d)
    i = 0
    while i < len_d:
        if d[i] != prev:
            e.append(d[i])
            prev = d[i]
        i += 1
    return e

my_list = ['a', 'z', 'a', 'b', 'c', 'z']
print(my_list)
print(remove_dups(my_list))

```

```python
# 10.15 Ex. 11 - word find using binary search or bisection search
import time
def bisect(t, w):
    """Returns index value of word w in sorted list t; or None if not found."""
    len_t = len(t)
    lower = 0
    upper = len_t
    i = (lower + upper) // 2
    count = 0
    while True:
        count += 1
        # print(count, lower, upper, i, w, t[i], w>t[i], w<t[i])
        if lower > upper:
            break
        if w > t[i]:
            lower = i +1
            i = (lower + upper) // 2
        elif w < t[i]:
            upper = i -1
            i = (lower + upper) // 2
        elif w == t[i]:
            return i
    
def build_sorted_list(file):
    """Returns sorted list from words.txt"""
    fin = open(file)
    words = []
    for line in fin:
        words.append(line.strip())
    fin.close()
    words.sort()
    return words

my_file = 'words.txt'
sorted_list = build_sorted_list(my_file)
my_word = input("Word to search for: ")
t_beg = time.time()
index = bisect(sorted_list, my_word)
if index == None:
    print(time.time()-t_beg, index)
else:
    print(time.time()-t_beg, index, sorted_list[index])
```

```python
# 11.1 Ex. 2 - histogram and get + 11.2 Ex. 3 using keys + 11.3 reverse lookup + 11.4 invert dict
# and setdefault()
def histogram(s):
    """Creates histogram of letter counts of string s, returns as dictionary d."""
    d = dict()
    for c in s:
        d[c] = d.get(c, 0) + 1
    return d

def print_hist(h):
    """use keys method to print keys and values alphabetically"""
    k = sorted(h.keys())
    for key in k:
        print(key, h[key])

def reverse_lookup(d, v):
    """Returns list of all keys in dict d having value v"""
    k_list = []
    for k in d:
        # print(k, v, type(v), d[k], type(d[k]))
        if d[k] == v:
            k_list.append(k)
    return k_list

def invert_dict(d):
    """Returns new dict, inverse, that maps values to keys and keys to a list of values"""
    inverse = dict()
    for key in d:
        val = d[key]
        inverse.setdefault(val, []).append(key)
#        single line with setdefault replaces next 4 lines
#        if val not in inverse:
#            inverse[val] = [key]
#        else:
#            inverse[val].append(key)
    return inverse

my_word = input("What is your word? ")
h = histogram(my_word)
print_hist(h)
my_v = int(input("Enter value to search for: "))
print("Reverse lookup: ", reverse_lookup(h, my_v))
print("Inverse lookup: ", invert_dict(h)[my_v])
```

```python
# 11.5 Memos - use to speed up fibonacci_old
import time
known = {0:0, 1:1}
def fibonacci_old(n):
    if not isinstance(n, int):
        return("Fibonacci only defined for integers.")
    if n < 0:
        return("Fibonacci only defined for positive integers.")
    elif n == 0:
        return(0)
    elif n == 1:
        return(1)
    else:
        return(fibonacci(n-1)+fibonacci(n-2))
    
def fibonacci(n):
    """Use memo, previously calculated value stored for later use, to speed up fibonacci calculation"""
    if n in known:
        return known[n]
    res = fibonacci(n-1) + fibonacci(n-2)
    known[n] = res
    return res
    
v = 0
t_beg1 = time.time()
while v < 1000:
    print(fibonacci_old(v))
    v += 1
t_end1 = time.time()

v = 0
t_beg2 = time.time()
while v < 150:
    print(fibonacci(v))
    v += 1
t_end2 = time.time()
print("Duration: (old) ", t_end1-t_beg1)
print("Duration: (new) ", t_end2-t_beg2)
```

```python
# 11.7 Ex. 9 has_duplicates improvement
"""This module contains code from
Think Python by Allen B. Downey
http://thinkpython.com

Copyright 2012 Allen B. Downey
License: GNU GPLv3 http://www.gnu.org/licenses/gpl.html

"""


def has_duplicates(t):
    """Checks whether any element appears more than once in a sequence.

    Simple version using a for loop.

    t: sequence
    """
    d = {}
    for x in t:
        if x in d:
            return True
        d[x] = True
    return False


def has_duplicates2(t):
    """Checks whether any element appears more than once in a sequence.

    Faster version using a set.

    t: sequence
    
    Detailed explanation (owa): set(t) makes a set from the list, t. Since sets can not include duplicates,
    len(set(t)) will be less than len(t) if t includes duplicates.
    """
    return len(set(t)) < len(t)


if __name__ == '__main__':
    t = [1, 2, 3]
    print(has_duplicates(t))
    t.append(1)
    print(has_duplicates(t))

    t = [1, 2, 3]
    print(has_duplicates2(t))
    t.append(1)
    print(has_duplicates2(t))
```

```python
"""
Think Python 12.9 structshape.py

This module provides one function, structshape(), which takes
an object of any type and returns a string that summarizes the
"shape" of the data structure; that is, the type, size and
composition.

This module is part of Swampy, a suite of programs available from
allendowney.com/swampy.

Copyright 2012 Allen B. Downey
Distributed under the GNU General Public License at gnu.org/licenses/gpl.html.

"""

from __future__ import print_function

def structshape(ds, pending=None):
    """Returns a string that describes the shape of a data structure.

    ds: any Python object

    Returns: string
    """
    if pending is None:
        pending = set()

    if id(ds) in pending:
        return 'infinity'

    pending = pending | set([id(ds)])

    typename = type(ds).__name__

    # handle sequences
    sequence = (list, tuple, set, frozenset, type(iter('')))
    if isinstance(ds, sequence):
        t = []
        for i, x in enumerate(ds):
            t.append(structshape(x, pending))
        rep = '%s of %s' % (typename, listrep(t))
        return rep

    # handle dictionaries
    elif isinstance(ds, dict):
        keys = set()
        vals = set()
        for k, v in ds.items():
            keys.add(structshape(k, pending))
            vals.add(structshape(v, pending))
        rep = '%s of %d %s->%s' % (typename, len(ds), 
                                   setrep(keys), setrep(vals))
        return rep

    # handle other types
    else:
        if hasattr(ds, '__class__'):
            return ds.__class__.__name__
        else:
            return typename


def listrep(t):
    """Returns a string representation of a list of type strings.

    t: list of strings

    Returns: string
    """
    if len(t) == 0:
        return 'empty'

    current = t[0]
    count = 0
    res = []
    for x in t:
        if x == current:
            count += 1
        else:
            append(res, current, count)
            current = x
            count = 1
    append(res, current, count)
    return setrep(res)


def setrep(s):
    """Returns a string representation of a set of type strings.

    s: set of strings

    Returns: string
    """
    rep = ', '.join(s)
    if len(s) == 1:
        return rep
    else:
        return '(' + rep + ')'
    return 


def append(res, typestr, count):
    """Adds a new element to a list of type strings.

    Modifies res.

    res: list of type strings
    typestr: the new type string
    count: how many of the new type there are

    Returns: None
    """
    if count == 1:
        rep = typestr
    else:
        rep = '%d %s' % (count, typestr)
    res.append(rep)


if __name__ == '__main__':
    t = []
    print(structshape(t))

    t = [1,2,3]
    print(structshape(t))

    t2 = [[1,2], [3,4], [5,6]]
    print(structshape(t2))

    t3 = [1, 2, 3, 4.0, '5', '6', [7], [8], 9]
    print(structshape(t3))

    class Point:
        """trivial object type"""

    t4 = [Point(), Point()]
    print(structshape(t4))

    s = set('abc')
    print(structshape(s))

    lt = zip(t, s)
    print(structshape(lt))

    d = dict(lt)        
    print(structshape(d))

    it = iter('abc')
    print(structshape(it))

    t = []
    t.append(t)
    print(structshape(t))
```

```python
#12.11 Ex 3 most_frequent
def most_frequent(s):
    """Takes a string, s and prints its letters in decreasing order of frequency"""
    t_r = []
    h = make_hist(s)
    for x in h:
        if x != '\n':
            t_r.append((h[x], x))
    t_r.sort(reverse=True)
    return t_r

    
def make_hist(s):
    d = dict()
    for c in s:
        d[c] = d.get(c, 0) + 1
    return d


def print_hist(h):
    """use keys method to print keys and values alphabetically"""
    k = sorted(h.keys())
    for key in k:
        print(key, h[key])

        
def open_file(filename):
    """opens file, filename and reads content, returns content as string."""
    return open(filename).read()

filename = 'words.txt'
s_file = open_file(filename)
d_r = most_frequent(s_file)
for freq, letter in d_r:
    print(freq, letter)
```
